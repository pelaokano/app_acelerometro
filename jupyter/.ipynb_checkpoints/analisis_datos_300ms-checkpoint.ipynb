```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
```

## Movimiento a la derecha, celular con pantalla frente a la persona

```python
data_derecha = pd.read_csv("../datos_300ms/movimiento_derecha.csv", header=None)
```

```python
data_derecha.head(10)
```

```python
data_derecha.iloc[:, 4] = data_derecha.iloc[:, 4].map(lambda x: int(x.replace("'", "")))
data_derecha.iloc[:, 0] = pd.to_datetime(data_derecha.iloc[:, 0].map(lambda x: str(x.replace("b'", ""))))
```

```python
data_derecha = data_derecha.loc[data_derecha[4]==1, :]
data_derecha = data_derecha.iloc[:, 1:4].values
```

```python
data_derecha.shape
```

```python
data_derecha = data_derecha.reshape(int(data_derecha.shape[0]/15), 15, data_derecha.shape[1])
```

```python
data_derecha.shape
```

```python
fig, axs = plt.subplots(10, 10, figsize=(20, 20))

for i in range(10):
    for j in range(10):
        curva = np.random.randint(0,105)
        axs[i, j].title.set_text(f'curva numero: {curva}')
        axs[i, j].grid()
        axs[i, j].plot(list(range(len(data_derecha[curva, :, :]))), data_derecha[curva, :, 0], 'r')
        axs[i, j].plot(list(range(len(data_derecha[curva, :, :]))), data_derecha[curva, :, 1], 'b')
        axs[i, j].plot(list(range(len(data_derecha[curva, :, :]))), data_derecha[curva, :, 2], 'g')
```

## Modelo

```python
data_adelante.shape
```

```python
etiqueta_adelante=np.array([0 for i in range(102)])
```

```python
etiqueta_adelante
```

```python
data_derecha.shape
```

```python
etiqueta_derecha=np.array([1 for i in range(105)])
```

```python
etiqueta_derecha
```

```python
data_izquierda.shape
```

```python
etiqueta_izquierda=np.array([2 for i in range(105)])
```

```python
etiqueta_izquierda
```

```python
data_X = np.concatenate((data_derecha, data_izquierda, data_adelante))
```

```python
data_y = np.concatenate((etiqueta_derecha, etiqueta_izquierda, etiqueta_adelante))
```

```python
data_y2 = np_utils.to_categorical(data_y)
```

```python
data_y2
```

```python
#etiquetas
#0 --> adelante
#1 --> derecha
#2 --> izquierda
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data_X, data_y2, test_size=0.2, random_state=42)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam

model = Sequential()
model.add(LSTM(50, input_shape=(15, 3)))
model.add(Dropout(0.5))
model.add(Dense(50, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
model.summary()
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_train, y_train, epochs=10, batch_size=1, verbose=2)
```

```python
accr = model.evaluate(X_test,y_test)
print('Test set\n  Loss: {:0.3f}\n  Accuracy: {:0.3f}'.format(accr[0],accr[1]))
```

```python
plt.title('metricas')
plt.plot(history.history['loss'], label='loss')
plt.plot(history.history['accuracy'], label='accuracy')
plt.legend()
plt.show();
```

```python
y_pred = model.predict(X_test)
```
